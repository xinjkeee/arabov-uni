```python

def is_pow_2(num: int) -> bool:
    while num > 1:
        num /= 2
    return num - int(num) == 0

n = int(input())
m = int(input())

a = [ [i for i in range(n)] for j in range(m) ]

print(a)

for i in range(n):
    for j in range(m):
        if is_pow_2(i+j):
            print(a[i-1][j-1], end=' ')
    print() 

```

```python
def count_words(s: str) -> dict:
    res = {ch: s.count(ch) for ch in s}
    return res
a = "aaasdf"
print(count_words(a))
```

```python
def merge_dicts(d1: dict, d2: dict) -> None:
    for key, val in d2.items():
        d1[key] = d1.get(key, val) + val
    return d1

d1 = {'a': 3, 's': 1, 'd': 1, 'f': 1}
d2 = {'a': 4, 's': 1, 'd': 1, 'f': 1}
print(merge_dicts(d1, d2))
```

```python
def get_min_key(d: dict) -> int:
    return min(d, key=d.get)
my_dict = {"a": 10, "b": 5, "c": 20}
print(get_min_key(my_dict))
```

```python
def search(d: dict, v: int) -> list:
    res = []
    for key, val in d.items():
        if val == v: res.append(key)
    return res
my_dict = {"a": 20, "b": 5, "c": 20}
val = 20
print(search(my_dict, val))
```

```python
def sort_dict(d: dict) -> dict:
    return dict(sorted(d.items(), key=lambda x: x[0]))

my_dict = {"a": 20, "c": 5, "b": 20}
print(sort_dict(my_dict))
```

```python
def sum_vals(d: dict) -> int:
    return sum(i for i in d.values())
my_dict = {"a": 20, "c": 5, "b": 20}
print(sum_vals(my_dict))
```

```python
def keys_odd(d: dict) -> list:
    return [key for key, val in d.items() if val % 2 == 1]
my_dict = {"a": 20, "c": 5, "b": 21}
print(keys_odd(my_dict))
```

```python
def diff_dict(d1: dict, d2: dict) -> dict:
    d1 = dict(set(d1.items()) - set(d2.items()))
    return d1
my_dict = {"a": 20, "c": 5, "b": 21, "be": 20}
my_dict2 = {"a": 20, "c": 5, "b": 21, "ae": 20}
print(diff_dict(my_dict, my_dict2))
```

```python
def max_keylen(d: dict):
    return max(d.keys(), key=len)
my_dict = {"a": 20, "c": 5, "b": 21, "be": 20}
print(max_keylen(my_dict))
```

```python
def mean_dict_vals(d: dict) -> float:
    return sum(d.values()) / len(d) if d else 0.0
my_dict = {"a": 20, "c": 5, "b": 21, "be": 20}
print(mean_dict_vals(my_dict))
```

```python
def deepcopy_dict(d: dict) -> dict:
    res = {}
    for key, val in d.items(): res[key] = val
    return res
my_dict = {"a": 20, "c": 5, "b": 21, "be": 20}
deepcopy = deepcopy_dict(my_dict)
deepcopy["a"] += 20
print(deepcopy)
print(my_dict)
```

```python
def set_from_list(l: list) -> set:
    return set(l)
my_set = {1, 1, 2, 3, 4}
print(set_from_list(my_set))
```

```python
def set_union(s1: set, s2: set) -> set:
    res = set()
    for val in s1: 
        if val in s2:
            res.add(val)
    return res
my_set = {1, 1, 2, 3, 4}
my_set2 = {1, 1, 2, 3, 4, 5}
print(set_union(my_set, my_set2))
```

```python
def is_subset(s1:set, s2:set) -> bool:
    for val in s1:
        if val not in s2:
            return False
    return True
my_set = {1, 2, 3, 4}
my_set2 = {1, 2, 3, 4, 5}
print(is_subset(my_set, my_set2))
```

```python
def del_elems(s: set, l: list) -> set:
    for elem in l:
        if elem in s:
            s.remove(elem)
    return s
my_set = {1, 2, 3, 4}
l = [2, 4]
print(del_elems(my_set, l))
```

```python
def symm_diff(l1: list, l2: list) -> list:
    return list(set(l1) ^ set(l2))
a = [1, 3, 5, 6]
b = [2, 4, 6]
print(symm_diff(a, b))
```

```python
def max_elem(s: set) -> int:
    if not s: return None
    max_elem = float("-inf")
    for elem in s: max_elem = max(max_elem, elem)
    return max_elem
my_set = {1, 2, 3, 4}
print(max_elem(my_set))
```

```python
def freq_elems(l: list) -> dict:
    unq_elem = set(l)
    freq = {}
    for elem in unq_elem:
        freq[elem] = l.count(elem)
    return freq
my_set = [1, 1, 2, 3, 4]
print(freq_elems(my_set))
```

```python
def divide_sets(s: set) -> tuple[set, set]:
    return {i for i in s if i % 2 == 0}, {i for i in s if i % 2 == 1}
my_set = {1, 1, 2, 3, 4}
print(divide_sets(my_set))
```

```python
def same_elem(s1: set, s2: set) -> bool:
    for elem in s1:
        if elem in s2: return True
    return False
my_set = {1, 1, 2, 3, 4}
my_set2 = {4, 6, 6}
print(same_elem(my_set, my_set2))
```

```python
def set_sorted_union(s1: set, s2: set) -> list:
    return sorted(list(s1 | s2))
my_set = {1, 1, 2, 3, 4}
my_set2 = {4, 6, 6}
print(set_sorted_union(my_set, my_set2))
```

```python
def mut_excl(s1: set, s2: set) -> bool:
    for elem in s1:
        if elem in s2: return False
    return True
my_set = {1, 1, 2, 3, 4}
my_set2 = {5, 6, 6}
print(mut_excl(my_set, my_set2))
```

```python
def tuple_from_list(l: list) -> tuple:
    return tuple(l)
my_list = [1, 1, 2, 3, 4]
print(tuple_from_list(my_list))
```

```python
def is_in_tuple(t: tuple, val: int) -> bool:
    return val in t
my_list = (1, 1, 2, 3, 4)
print(is_in_tuple(my_list, 1))
```

```python
def max_elem(t: tuple) -> int:
    mx = float("-inf")
    for elem in t: mx = max(mx, elem)
    return mx
my_list = (1, 1, 2, 3, 4)
print(max_elem(my_list))
```

```python
def divide_tuple(t: tuple) -> tuple[tuple, tuple]:
    t1 = t[:len(t)//2]
    t2 = t[len(t)//2:]
    return (t1, t2)
my_tuple = (1, 1, 2, 3, 4, 6)
print(divide_tuple(my_tuple))
```

```python
def conc_tuples(l: list[tuple]) -> tuple:
    return tuple(elem for t in l for elem in t)
my_list = [(1, 1, 2, 3, 4, 6), (1, 19)]
print(conc_tuples(my_list))
```

```python
def filter_tuple(t: tuple, tol: int) -> tuple:
    return tuple([val for val in t if val > tol])
my_tuple = (1, 1, 2, 3, 4, 6)
print(filter_tuple(my_tuple, 1))
```

```python
def str_to_tuple(s: str) -> tuple:
    return tuple(i for i in s)
my_str = "good day"
print(str_to_tuple(my_str))
```

```python
def nested_tuples() -> tuple:
    letters = ['a', 'b', 'c', 'd', 'e']
    return tuple((letter, number) for letter in letters for number in range(1, 6))
print(nested_tuples())
```

```python
def tuple_to_str(t: tuple) -> str:
    return " ".join(t)
t = ("hello", "world")
print(tuple_to_str(t))
```

```python
def tuple_odd(rng: tuple[int, int]) -> tuple:
    return tuple(i for i in range(rng[0], rng[1]) if i % 2 == 1)
rng = (2, 10)
print(tuple_odd(rng))
```

```python
def group_elems(t: tuple) -> dict:
    list_types = [type(i) for i in set(t)]
    d = {i: [] for i in list_types}
    for elem in t:
        d[type(elem)].append(elem)
    return d

t = ("asd", "asdsa", 1, 2, 3.0, 1232, 3.123, "asd")
print(group_elems(t))
```
