```python
### 3 
roman = {
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'M': 1000
}

num = "MMMMMMCCLXXIV"

def roman_to_natural(num: str) -> int:
    res = 0
    pred_char = ""
    for char in num:
        res += roman[char]
        if pred_char == "I" and char != "I":
            res -= 2
        pred_char = char
    return res
roman_to_natural(num)
```

```python
### 1
from random import randint
import random
from math import sqrt
random.seed(42)
with open("file1.txt", "w") as f:
    for i in range(100):
        random_number = randint(1, 1000000)
        f.write(str(random_number))
        f.write(str(" "))
with open("file1.txt", "r") as f:
    a = f.read()
    a = [int(i) for i in a.split()]
count = 0
for num in a:
    if (sqrt(num) % 1 == 0) and (num % 2 != 0):
      count+=1
print(count)
```

```python
## 2
import pandas as pd


def make_assort():
    toy_names = ["LEGO", "Ninjago", "Transformer", "Barbie Doll"]

    data = []
    for i in range(len(toy_names)):
        age = random.choice([2, 5, 11])
        price = random.randint(10000, 20000)
        toy = random.choice(toy_names)
        data.append([toy, price, age])

    df = pd.DataFrame(data, columns=["Name", "Price", "Age"])
    df.to_csv("Assort.csv")

make_assort()
assort = pd.read_csv("Assort.csv")
assort[(assort.Price <= 14000) & ((assort.Age == 5) | (assort.Age == 11))]
# assort
```

```python
import random
import datetime

QUESTION_FILE = 'questions.txt'
RESULT_FILE = 'results.txt'

def load_questions(filename):
    questions = []
    try:
        with open(filename, encoding='utf-8') as f:
            for line in f:
                parts = line.strip().split('|')
                if len(parts) != 7:
                    print(f"Неверный формат строки: {line.strip()}")
                    continue
                question_text = parts[0]
                options = parts[1:6]
                correct = parts[6]
                questions.append({
                    'question': question_text,
                    'options': options,
                    'correct': correct
                })
    except FileNotFoundError:
        print(f"Файл {filename} не найден.")
        exit(1)
    return questions

def shuffle_question(q):
    options = q['options'][:]
    random.shuffle(options)
    correct_index = options.index(q['correct']) + 1  # 1-based index
    return q['question'], options, correct_index

def ask_questions(questions):
    correct_count = 0
    total = len(questions)
    for i, q in enumerate(questions, 1):
        question_text, options, correct_index = shuffle_question(q)
        print(f"\nВопрос {i}/{total}: {question_text}")
        for idx, opt in enumerate(options, 1):
            print(f"{idx}. {opt}")
        while True:
            try:
                answer = int(input("Ваш ответ (1-5): "))
                if 1 <= answer <= 5:
                    break
                else:
                    print("Введите число от 1 до 5.")
            except ValueError:
                print("Пожалуйста, введите число.")
        if answer == correct_index:
            print("Правильно!")
            correct_count += 1
        else:
            print(f"Неправильно. Правильный ответ: {correct_index}. {options[correct_index - 1]}")
    return correct_count, total

def save_results(start_time, end_time, correct, total):
    percent = (correct / total) * 100
    with open(RESULT_FILE, 'w', encoding='utf-8') as f:
        f.write(f"Время начала теста: {start_time}\n")
        f.write(f"Время окончания теста: {end_time}\n")
        f.write(f"Общее количество вопросов: {total}\n")
        f.write(f"Количество правильных ответов: {correct}\n")
        f.write(f"Процент правильных ответов: {percent:.2f}%\n")
    print(f"\nРезультаты сохранены в файл {RESULT_FILE}")

def main():
    questions = load_questions(QUESTION_FILE)
    if not questions:
        print("Нет допустимых вопросов.")
        return
    random.shuffle(questions)
    start_time = datetime.datetime.now()
    correct, total = ask_questions(questions)
    end_time = datetime.datetime.now()
    print("\nТестирование завершено.")
    print(f"Правильных ответов: {correct} из {total} ({(correct / total) * 100:.2f}%)")
    save_results(start_time, end_time, correct, total)

if __name__ == '__main__':
    main()
```
